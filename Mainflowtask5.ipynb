```python
import tkinter as tk
from tkinter import ttk
import requests
from PIL import ImageTk, Image
import json

def convert_currency():
    """Converts the currency based on user input."""
    source = from_currency_combo.get()
    destination = to_currency_combo.get()
    amount = amount_entry.get()

    # Send a request to the Pair Conversion url and convert it to json
    result = requests.get(f'https://v6.exchangerate-api.com/v6/{API_KEY}/pair/{source}/{destination}/{amount}').json()

    # Get the conversion result from response result
    converted_result = result['conversion_result']

    # Format the results
    formatted_result = f'{amount} {source} = {converted_result} {destination}'

    # Add text to the empty result label
    result_label.config(text=formatted_result)

    # Add text to the empty time label
    time_label.config(text='Last updated,' + result['time_last_update_utc'])

def clear_fields():
    """Clears the input fields and result labels."""
    amount_entry.delete(0, tk.END)
    result_label.config(text="")
    time_label.config(text="")

# Main window setup
window = tk.Tk()
window.geometry('600x300+500+200')
window.title('Currency Converter')
window.iconbitmap('C:/Users/DELL/Downloads/icon.ico')
window.resizable(False, False)

# Load and resize image
image = Image.open('C:/Users/DELL/Downloads/currency.png')
zoom = 0.5
resized_image = image.resize((int(image.width * zoom), int(image.height * zoom)))
photo = ImageTk.PhotoImage(resized_image)

# The top frame
top_frame = tk.Frame(window, width=300, height=80)
top_frame.grid(row=0, column=0)

# Label for the text Currency Converter
name_label = tk.Label(top_frame, text='USD Currency Converter Using Python', pady=30, padx=24, justify=tk.CENTER, font=('Poppins 20 bold'))
name_label.grid(row=0, column=0)

# The bottom frame
bottom_frame = tk.Frame(window, width=300, height=250)
bottom_frame.grid(row=1, column=0)

# Widgets inside the bottom frame
from_currency_label = tk.Label(bottom_frame, text='FROM:', font=('Poppins 10 bold'), justify=tk.LEFT)
from_currency_label.place(x=5, y=10)
to_currency_label = tk.Label(bottom_frame, text='TO:', font=('Poppins 10 bold'), justify=tk.RIGHT)
to_currency_label.place(x=160, y=10)

# Combobox for holding from_currencies
from_currency_combo = ttk.Combobox(bottom_frame, width=14, font=('Poppins 10 bold'))
from_currency_combo.place(x=5, y=30)

# Combobox for holding to_currencies
to_currency_combo = ttk.Combobox(bottom_frame, width=14, font=('Poppins 10 bold'))
to_currency_combo.place(x=160, y=30)

# Label for AMOUNT
amount_label = tk.Label(bottom_frame, text='AMOUNT:', font=('Poppins 10 bold'))
amount_label.place(x=5, y=55)

# Entry for amount
amount_entry = tk.Entry(bottom_frame, width=25, font=('Poppins 15 bold'))
amount_entry.place(x=5, y=80)

# Empty label for displaying the result
result_label = tk.Label(bottom_frame, text='', font=('Poppins 10 bold'), justify=tk.RIGHT)
result_label.place(x=5, y=115)

# Empty label for displaying the time
time_label = tk.Label(bottom_frame, text='', font=('Poppins 10 bold'))
time_label.place(x=5, y=135)

# The clickable button for converting the currency
convert_button = tk.Button(bottom_frame, text="SEARCH",  bg="red", fg="white", font=('Poppins 10 bold'), justify=tk.LEFT, command=convert_currency)
convert_button.place(x=20, y=170)

clear_button = tk.Button(bottom_frame, text="CLEAR",  bg="blue", fg="white", font=('Poppins 10 bold'), justify=tk.RIGHT, command=clear_fields)
clear_button.place(x=180, y=170)

API_KEY ='1d9a7439e359b82b63836095'
# The Standard request url
url = f'https://v6.exchangerate-api.com/v6/{API_KEY}/latest/USD'
# Making the Standard request to the API and converting the request to json
response = requests.get(f'{url}').json()
# Converting the currencies to dictionaries
currencies = dict(response['conversion_rates'])

# Populate comboboxes with currencies
from_currency_combo['values'] = list(currencies.keys())
from_currency_combo.current(0)
to_currency_combo['values'] = list(currencies.keys())

window.mainloop()
```
